```python
import os
import cv2
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import pickle

DATASET_PATH = "/content"

features = []
labels = []

for file in os.listdir(DATASET_PATH):

    if not file.lower().endswith(".png"):
        continue

    file_path = os.path.join(DATASET_PATH, file)

    img = cv2.imread(file_path)
    if img is None:
        continue

    img = cv2.resize(img, (100, 100))

    avg_color = np.mean(img, axis=(0, 1))

    # Extract label from filename
    # example: red_1.png → "red"
    label = file.split("_")[0].lower()

    features.append(avg_color)
    labels.append(label)

# Convert to numpy
X = np.array(features)
y = np.array(labels)

# Safety check
if len(X) == 0:
    raise ValueError("❌ No images were loaded. Make sure PNG files exist in /content.")

# Train model
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)

# Save model
with open("color_model.pkl", "wb") as f:
    pickle.dump(model, f)

print("✅ Model trained successfully!")
print("Classes detected:", set(labels))
```

```python
import os
import cv2
import numpy as np
import pickle

image_path = "/content/PUT_YOUR_IMAGE_NAME_HERE.png"

if not os.path.exists(image_path):
    raise FileNotFoundError(f"❌ File not found: {image_path}")

with open("color_model.pkl", "rb") as f:
    model = pickle.load(f)

img = cv2.imread(image_path)
img = cv2.resize(img, (100, 100))

avg_color = np.mean(img, axis=(0, 1))
avg_color = avg_color.reshape(1, -1)

prediction = model.predict(avg_color)

print("✅ Predicted color:", prediction[0])
```
